```python
#Back Propagation
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42) # to ensure consistent numbers
```

```python
W1=np.random.rand(2,2)
W2=np.random.rand(2,1)
I2=np.zeros(2)
Y2=np.zeros(2)
I3=np.zeros(1)
Y3=np.zeros(1)
```

```python
I1=np.array([0.5,0.5]) # Input Vector
```

```python
def sigmoid(arr):
    return 1/(1+np.exp(-arr))
```

```python
def forward(I1, W1, W2):
    # Hidden Layer
    I2 = np.dot(I1, W1)
    Y2 = sigmoid(I2)
    # Output Layer
    I3 = np.dot(Y2, W2)
    Y3 = sigmoid(I3)
    return Y2, Y3
```

```python
def backprop(I1, Y2, Y3, target, W1, W2, learning_rate=0.1):
    error_output = Y3 - target
    delta3 = error_output * (Y3 * (1 - Y3))
    error_hidden = np.dot(delta3, W2.T)
    delta2 = error_hidden * (Y2 * (1 - Y2))
    I1_2d = np.atleast_2d(I1)
    Y2_2d = np.atleast_2d(Y2)
    W2_grad = np.dot(Y2_2d.T, np.atleast_2d(delta3))
    W1_grad = np.dot(I1_2d.T, np.atleast_2d(delta2))
    W2 -= learning_rate * W2_grad
    W1 -= learning_rate * W1_grad
    return W1, W2, error_output
```

```python
target = 0.8
epochs = 1000
lr=0.1
for i in range(epochs):
    Y2, Y3 = forward(I1, W1, W2)
    W1, W2, error = backprop(I1, Y2, Y3, target, W1, W2, learning_rate=lr
    if i % 100 == 0:
        loss = 0.5 * np.square(error)
        print(f"Epoch {i}: Loss = {loss[0]:.6f}, Output = {Y3[0]:.4f}")
print(f"\nFinal Prediction: {Y3[0]:.4f}")
```
