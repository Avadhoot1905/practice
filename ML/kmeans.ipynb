```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
path="./loan_data.csv"
df=pd.read_csv(path)
df.info()
```

```python
df=df.drop(labels='purpose',axis=1)
kmeans=KMeans(n_clusters=4)
model=kmeans.fit(df)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```
