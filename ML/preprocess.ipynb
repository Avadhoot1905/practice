```python
from pandas import read_csv
from numpy import set_printoptions
from sklearn import preprocessing
from sklearn.preprocessing import Normalizer, Binarizer, StandardScaler
```

```python
names = ['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe = read_csv("/home/anaconda/dataset2/pima-indians-diabetes.csv")
print(dataframe)
```

```python
array = dataframe.values
data_scalar = preprocessing.MinMaxScaler(feature_range=(0,1))
print(data_scalar)
```

```python
data_rescaled = data_scalar.fit_transform(array)
print(data_rescaled)
```

```python
set_printoptions(precision = 2)
print("\nScaled data:\n", data_rescaled[0:10])
print("Scaling Complete")
```

```python
Data_normaliser = Normalizer(norm = 'l2').fit(array)
Data_normalised = Data_normaliser.fit_transform(array)
print(Data_normalised)
```

```python
set_printoptions(precision = 2)
print("\nNormalised Data:\n",Data_normalised[:3])
print("Normalisation complete")
```

```python
bina = Binarizer(threshold=0.5).fit(array)
DB = bina.fit_transform(array)
print(DB)
```

```python
set_printoptions(precision = 2)
print("\nBinary Data:\n", DB[0:3])
print("Binarisation complete")
```

```python
ds = StandardScaler().fit(array)
dr = ds.transform(array)
print(dr)
```

```python
set_printoptions(precision=2)
print("\nStandardised data:\n",dr[:4])
print("Standardisation complete")
```
